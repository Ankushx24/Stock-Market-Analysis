```python

import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_csv('/content/stock_data.csv', parse_dates=['Date'])
    df.set_index('Date', inplace=True)


    df['Daily_Return'] = df['Close'].pct_change()


    plt.figure(figsize=(12, 6))

    plt.subplot(2, 1, 1)
    plt.plot(df['Close'])
    plt.title('Stock Closing Price')
    plt.ylabel('Price')

    plt.subplot(2, 1, 2)
    plt.plot(df['Daily_Return'])
    plt.title('Daily Returns')
    plt.ylabel('Return')

    plt.tight_layout()
    plt.show()

    print(df.describe())


except FileNotFoundError:
    print("Error: File not found. Please make sure the file exists and the path is correct.")
except KeyError as e:
    print(f"Error: Column '{e}' not found in the CSV file. Please check your file.")
except Exception as e:
    print(f"An error occurred: {e}")
```
